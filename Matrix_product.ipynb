```python
def mat_dot(a_array, b_array):
  row_size = a_array.shape[0]
  col_size = a_array.shape[1]
  dot_pro = np.array([])
  dotx = 0
  for i in range(row_size):
    for j in range(row_size):
      doty = 0
      for k in range(col_size):
        doty = doty + a_array[i,k]*b_array[k,j]
      dot_pro = np.append(dot_pro, doty)
  product = dot_pro.reshape(row_size,col_size)
  return product
```

```python
dot_product = mat_dot(a, b)
print(dot_product)
```
