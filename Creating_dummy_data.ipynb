# [Problem 2] Visualization by scatter plot

```python
import matplotlib.pyplot as plt
mean1 = (-3, 0)
mean = (0,3)
cov = [[1.0, 0.8],[0.8, 1.0]]
np.random.seed(0)

x = np.random.multivariate_normal(mean1, cov, size = (500))
y = np.random.multivariate_normal(mean, cov, size = (500))
print(x.shape)
plt.scatter(x[:,0],x[:,1],marker='x')
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

# [Problem 4] Addition of data

```python
mean1 = (-3, 0)
mean = (0 , -3)
cov = [[1.0, 0.8],[0.8, 1.0]]
np.random.seed(0)
x, y = np.random.multivariate_normal(mean1, cov, size = (500,1)).T
x1, y1 = np.random.multivariate_normal(mean, cov, size = (500,1)).T

fig, ax = plt.subplots()
ax.scatter(x,y,marker='x')
ax.scatter(x1,y1,marker='*',color='orange')
ax.set_title('scatter')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.legend(['0','1'])
plt.show()
```

# [Problem 5] Data combination

```python
np.random.seed(0)
x1 = np.random.multivariate_normal(mean1, cov, size = (500))
x2 = np.random.multivariate_normal(mean, cov, size = (500))
new = np.concatenate([x1, x2])
new
```

# [Problem 6] Labeling

```python
label1= np.zeros((500,1))
label2= np.ones((500,1))
x1_new = np.append(x1, label1,axis=1)
x2_new = np.append(x2, label2,axis=1)

new = np.concatenate([x1_new, x2_new])
new
```
