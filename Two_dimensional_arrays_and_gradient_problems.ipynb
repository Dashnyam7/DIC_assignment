## [Problem 1] Linear function

```python
import numpy as np
x = np.arange(-50,50.1,0.1).reshape((1001,1))
x
```

```python
y = x/2+1
y
```

## [Problem 2] Concatenation of arrays

```python
concat = np.concatenate((x, y),axis=1)
concat
```

## [Problem 3] Find the gradient

```python
xGrad = x[1:] - x[:-1]
yGrad = y[1:] - y[:-1]
gradient = yGrad/xGrad
gradient
```

## [Problem 4] Draw a graph

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x1 = np.arange(-50,50,0.1).reshape((1000,1))
ax.plot(x1,gradient)
plt.show()
```

## [Problem 5] Python functionalization

```python
def function1(array_x=np.arange(-50,50.1,0.1).reshape((1001,1))):
  array_y = array_x**2
  pass
  return array_y
```

```python
def function2(array_x=np.arange(-50,50.1,0.1).reshape((1001,1))):
  array_y = 2*array_x**2+2**array_x
  pass
  return array_y
```

```python
def function3(array_x=np.arange(0, 50.1, 0.1).reshape((501,1))):
  array_y = np.sin(np.sqrt(array_x))
  pass
  return array_y
```

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x_range : tuple
    x = np.arange(*x_range)
    x_size = x.size
    array_x = x.reshape((x_size,1))
    array_xy = np.concatenate((array_x,function), axis=1)
    gradient = (array_x[1:]-array_x[:-1])/(function[1:]-function[:-1])
    pass
    return array_xy, gradient
```

```python
array_xy, gradient1 = compute_gradient(function1())
print(array_xy)
print(gradient1)
```

```python
fig, ax = plt.subplots()
x1 = np.arange(-50,50,0.1).reshape((1000,1))
ax.plot(x1,gradient1)
plt.show()
```

```python
array_xy2, gradient2 = compute_gradient(function2())
fig, ax = plt.subplots()
x2 = np.arange(-50,50,0.1).reshape((1000,1))
ax.plot(x2,gradient2)
plt.show()
```

```python
array_xy3, gradient3 = compute_gradient(function3(),x_range=(0, 50.1, 0.1))
fig, ax = plt.subplots()
x3 = np.arange(0,50,0.1).reshape((500,1))
ax.plot(x3,gradient3)
plt.show()
```
